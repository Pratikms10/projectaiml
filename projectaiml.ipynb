```python
# Convert string labels to numeric (only needs to be done once)
y_train = y_train.replace({'<=50K': 0, '>50K': 1})
y_test = y_test.replace({'<=50K': 0, '>50K': 1})
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

# SVR pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svr', SVR())
])

# Fit the model
pipeline.fit(x_train, y_train)

# Predict
predict1 = pipeline.predict(x_test)

# Threshold predictions to get class labels
predict1_class = [1 if y >= 0.5 else 0 for y in predict1]

# Evaluate
accuracy = accuracy_score(y_test, predict1_class)
print("SVR Accuracy (with thresholding):", accuracy)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Make sure target is numeric
y_train_nb = y_train.replace({'<=50K': 0, '>50K': 1})
y_test_nb = y_test.replace({'<=50K': 0, '>50K': 1})

# Create and train Naive Bayes model
nb = GaussianNB()
nb.fit(x_train, y_train_nb)

# Predict and evaluate
predict_nb = nb.predict(x_test)
accuracy_nb = accuracy_score(y_test_nb, predict_nb)
print("Naive Bayes Accuracy:", accuracy_nb)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Make sure target is numeric
y_train_bag = y_train.replace({'<=50K': 0, '>50K': 1})
y_test_bag = y_test.replace({'<=50K': 0, '>50K': 1})

# âœ… Updated: Use estimator=... instead of base_estimator
bagging = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=100, random_state=42)
bagging.fit(x_train, y_train_bag)

# Predict and evaluate
predict_bag = bagging.predict(x_test)
accuracy_bag = accuracy_score(y_test_bag, predict_bag)
print("Bagging Classifier Accuracy:", accuracy_bag)

```

```python
from sklearn.ensemble import AdaBoostClassifier

# Create and train AdaBoost Classifier
adaboost = AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=1), n_estimators=100, random_state=42)
adaboost.fit(x_train, y_train_bag)

# Predict and evaluate
predict_ada = adaboost.predict(x_test)
accuracy_ada = accuracy_score(y_test_bag, predict_ada)
print("AdaBoost Classifier Accuracy:", accuracy_ada)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVR
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Ensure y is numeric
y_train_mod = y_train.replace({'<=50K': 0, '>50K': 1})
y_test_mod = y_test.replace({'<=50K': 0, '>50K': 1})

# Define models (SVR will be handled slightly differently)
models = {
    "KNN": KNeighborsClassifier(),
    "SVR": SVR(),  # Regression model
    "MLPClassifier": MLPClassifier(max_iter=2000, random_state=42),
    "GaussianNB": GaussianNB(),
    "BaggingClassifier": BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=100, random_state=42),
    "AdaBoostClassifier": AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=1), n_estimators=100, random_state=42)
}

results = {}

for name, model in models.items():
    # SVR needs separate handling since it's a regression model
    if name == "SVR":
        pipe = Pipeline([
            ('scaler', StandardScaler()),
            ('model', model)
        ])
        pipe.fit(x_train, y_train_mod)
        ypred_continuous = pipe.predict(x_test)
        ypred = [1 if y >= 0.5 else 0 for y in ypred_continuous]
    else:
        pipe = Pipeline([
            ('scaler', StandardScaler()),
            ('model', model)
        ])
        pipe.fit(x_train, y_train_mod)
        ypred = pipe.predict(x_test)

    acc = accuracy_score(y_test_mod, ypred)
    results[name] = acc
    print(f"\n{name} Accuracy: {acc:.4f}")
    print(classification_report(y_test_mod, ypred))
```

```python
import matplotlib.pyplot as plt

# Sort results by accuracy
sorted_results = dict(sorted(results.items(), key=lambda item: item[1]))

plt.figure(figsize=(10, 6))
plt.barh(list(sorted_results.keys()), list(sorted_results.values()), color='mediumseagreen')
plt.xlabel('Accuracy Score')
plt.title('Model Comparison (Sorted)')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVR
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import joblib

# ðŸ”„ Convert target to numeric if not already
y = y.replace({'<=50K': 0, '>50K': 1})

# ðŸ”€ Train-test split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# ðŸ§  Define models
models = {
    "KNN": KNeighborsClassifier(),
    "SVR": SVR(),  # Regression model - thresholded
    "MLPClassifier": MLPClassifier(max_iter=2000, random_state=42),
    "GaussianNB": GaussianNB(),
    "BaggingClassifier": BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=100, random_state=42),
    "AdaBoostClassifier": AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=1), n_estimators=100, random_state=42)
}

results = {}

# ðŸ” Train, Predict, Evaluate
for name, model in models.items():
    # Use a pipeline with StandardScaler for consistent scaling
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])

    # Fit model
    pipe.fit(X_train, y_train)

    # Predict
    if name == "SVR":
        y_pred = pipe.predict(X_test)
        y_pred = [1 if pred >= 0.5 else 0 for pred in y_pred]
    else:
        y_pred = pipe.predict(X_test)

    # Accuracy
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"{name}: {acc:.4f}")

# ðŸ† Select and Save the Best Model
best_model_name = max(results, key=results.get)
best_model_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', models[best_model_name])
])
best_model_pipeline.fit(X_train, y_train)

print(f"\nâœ… Best model: {best_model_name} with accuracy {results[best_model_name]:.4f}")

# ðŸ’¾ Save the best model
joblib.dump(best_model_pipeline, "best_model.pkl")
print("âœ… Saved best model as best_model.pkl")
```
